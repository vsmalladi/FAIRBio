# Array of Versions

```python
import json
versions = []
with open('biocontainers_software.json') as f:
    tools = json.load(f)
    for tool in tools.get("tools", []):
        for v in tool.get("versions", []):
            versions.append(v)
```

# Array of unique software names

```python
software = []
with open('biocontainers_software.json') as f:
    tools = json.load(f)
    for tool in tools.get("tools", []):
        software.append(tool.get("id"))
```
